```python
import random

from IPython.display import display

import newaresql
```

You must either set the enviroment variables
- `BTS_HOST`
- `BTS_PORT`
- `BTS_USER`
- `BTS_PASSWORD`
- `BTS_DATABASE`
  
or modify the credentials dictionary

```python
credentials = dict(host=None, port=None, user=None, password=None, database=None)

with newaresql.connect(**credentials) as conn:
    # conn.get_tests() will give all the tests metadata as a polars dataframe
    tests = conn.get_tests()
    print(f"Found {len(tests)} tests in the database")

    # newaresql.list_tests(connector=conn) pipes conn.get_tests().to_dicts()
    # and returns a list of dicts, each representing one test

    print("Selecting a random test...")
    test = random.choice(tests.to_dicts())

    print("Selected test:")
    print("\n".join(f"\t{k}: {v}" for k, v in test.items()))

    print(f"Fetching the first 1000 records")
    data = newaresql.get_data(test, connector=conn, where={"seq_id": (None, 1000)})
```
