```python
import pprint
import random

from IPython.display import display

import newaresql
```

```python
credentials = dict(
    host=...,
    port=...,
    user=...,
    password=...,
    database=...,
)
with newaresql.connect(**credentials) as conn:
    tests = newaresql.list_tests(connector=conn)
    print(f"Found {len(tests)} tests in the database")
    test = random.choice(tests)
    print(f"Selected test:")
    print("\n".join(f"\t{k}: {v}" for k, v in test.items()))
    data = newaresql.get_data(test, connector=conn, where={"seq_id": (None, 1000)})
    display(data)
```
